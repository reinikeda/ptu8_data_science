```python
# Nuskaitykite į DF faila
fire_data = pd.read_csv('text/top_20_CA_wildfires.csv')
fire_data.head()
```

```python
# Surūšiuokite eilutes pagal metus. Išsiaiškinkite, kaip rūšiuoti, kad naujausi gaisrai būtų viršuje
fire_data.sort_values('year', ascending=False)
```

```python
# Parašykite funkciją, kuri mėnesio pavadinimą verstų skaičiumi.
def month_converter(month):
    months = {1: 'January',
    2: 'February',
    3: 'March',
    4: 'April',
    5: 'May',
    6: 'June',
    7: 'July',
    8: 'August',
    9: 'September',
    10: 'October',
    11: 'November',
    12: 'December'}
    for num, name in months.items():
        if name == month:
            return num
```

```python
# Naudodami .apply() lentelėje pakeiskite mėnesių pavadinimus skaičiais.
fire_data['month'] = fire_data['month'].apply(month_converter)
```
