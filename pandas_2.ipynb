```python
import numpy as np
import pandas as pd
```

### trukstami duomenys

```python
df = pd.DataFrame(np.random.rand(5,6),
    ['a', 'b', 'c', 'd', 'e'],
    ['U', 'V', 'W', 'X', 'Y', 'Z'])
df
```

```python
table = df[df > 0.15]
table
```

```python
#  grazina nulines NaN reiksmes
table.isnull()
```

```python
# parodo kiek stulpelyje yra NaN reiksmiu
table.isnull().sum()
```

```python
# NaN reiksmiu tikrinimas konkreciame stulpelyje
table['V'].isnull()
```

```python
table['V'].isnull().sum()
```

### dropna()

```python
# ismeta eilutes, kuriose yra bent viena NaN reiksme. Palieka tik pilnai uzpildytas eilutes.
table.dropna()
```

```python
# ismeta stulpelius, kuriuose yra bent viena NaN reiksme.
table.dropna(axis=1)
```

```python
# ismeta eilutes, kuriose yra maziau negu nurodytos (5) uzpildytos reiksmes.
table.dropna(thresh=5)
```

```python
table.dropna(axis=1, thresh=5)
```

### fillna()

```python
#  uzpildo NaN reiksmes
table.fillna('antradienis')
```

```python
# uzpildo stilpelio vidurkiu
table['X'].fillna(value=table['X'].mean())
```

```python
# uzpildo visa lentele jos vidurkiu
table.fillna(value=table.mean())
```

## grupavimas

```python
duomenys = {'Šalis': ['Lietuva',
  'Lietuva',
  'Lietuva',
  'Latvija',
  'Latvija',
  'Latvija',
  'Estija',
  'Estija',
  'Estija'],
 'Miestas': ['Vilnius',
  'Kaunas',
  'Klaipėda',
  'Ryga',
  'Ventspilis',
  'Daugpilis',
  'Talinas',
  'Tartu',
  'Pernu'],
 'Gyv': [541, 287, 147, 716, 43, 105, 400, 101, 46]}
```

```python
data = pd.DataFrame(duomenys)
data
```

```python
baltic = data.groupby('Šalis')
baltic
```

```python
baltic.mean()
```

```python
baltic.sum()
```

```python
# suskaiciuoja eiluciu kieki, ne ju reiksmes, nes pvz gali buti string
baltic.count()
```

```python
baltic.max()
```

```python
baltic.min()
```

```python
# istraukiame konkrecios eilutes reiksme
baltic.sum().loc['Lietuva']
```

```python
# grazina pagrindine lenteles informacija
# -> mean, standard deviation, Q0(min), Q1, Q2(median), Q3, Q4(max)
baltic.describe()
```

```python
# lenteles pavertimas
baltic.describe().transpose()
```

```python
# konkretaus stulpelio statistiniai duomenys
baltic.describe().transpose()['Estija']
```

## DataFrame jungimas

```python
data
```

```python
# grazina iki 6 eilutes
data_main = data[0:6]
data_main
```

```python
# grazina nuo 6 eilutes
data_bottom = data[6:]
data_bottom
```

```python
pd.concat([data_main, data_bottom])
```

```python
pd.concat([data_bottom, data_main])
```

```python
# prijungimas is desines
pd.concat([data_main, data_bottom], axis=1)
```

```python
data_right = data_bottom.reset_index()
data_right.drop(columns='index', inplace=True)
data_right
```

```python
pd.concat([data_main, data_right], axis=1)
```

```python
lenkija = {'Šalis': ['Lenkija', 'Lenkija', 'Lenkija'], 
           'Miestas':['Varšuva', 'Vroclavas', 'Gdanskas'], 
          'Gyv': [1688, 638, 461]}
```

```python
pl = pd.DataFrame(lenkija)
data_right_1 = pd.concat([data_right, pl])
data_right = data_right_1.reset_index().drop(columns='index')
data_right
```

```python
pd.concat([data_main, data_right], 1)
```

```python
# sukuriamas bendras stulpelis
key = 'A B C D E F'.split()
data_main['key'] = key
data_right['key'] = key
```

```python
data_main
```

```python
data_right
```

```python
# sujungiama per bendro stulpelio reiksmemis
pd.merge(data_main, data_right, on='key')
```

```python
# sukuriamas dar vienas bendras stulpelis
key2 = 'B C D E F G'.split()
data_main['key2'] = key2
data_right['key2'] = key2
```

```python
data_main
```

```python
data_right
```

```python
# lenteliu sujungimas per 2 on stulpelius
pd.merge(data_main, data_right, on=['key', 'key2'])
```

```python
keyX = 'W C D E F W'.split()
data_right['key2'] = keyX
```

```python
data_right
```

```python
# sujungia tik tas eilutes, kuriu on reiksmes sutampa
pd.merge(data_main, data_right, on=['key', 'key2'])
```

```python
data_main.join(data_right, lsuffix='L', rsuffix='R')
```
