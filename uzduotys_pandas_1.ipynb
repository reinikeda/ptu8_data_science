```python
# Ištrinkite numeracijos stulpelį.
failas.drop('Numeracija', axis=1, inplace=True)
failas
```

```python
# Kurių miestų dar nebuvo 1959m.?
failas[failas['1959'] == 0]
```

```python
# Kokie miestai 1897 turėjo daugiau gyventojų, negu 2019?
failas[(failas['1897']) > (failas['2019'])]
```

```python
# Kuriuose miestuose padaugėjo gyventojų nuo 2011 iki 2019?
failas[(failas['2011']) < (failas['2019'])][['2011', '2019']]
```

```python
# Kuriuose miestuose gyventojų skaičius nuosekliai mažėjo nuo pat 1897m.?
failas[
    (failas['1897'] > failas['1923']) & 
    (failas['1923'] > failas['1959']) & 
    (failas['1959'] > failas['1970']) & 
    (failas['1970'] > failas['1979']) &
    (failas['1979'] > failas['1989']) &
    (failas['1989'] > failas['2001']) &
    (failas['2001'] > failas['2011']) &
    (failas['2011'] > failas['2019'])]
```

```python
skirtumas = (((failas['2019'] - failas['1989']) / failas['1989']) * 100)
```

```python
print(skirtumas.idxmax(), (skirtumas.max()).round(2), '%')
print(skirtumas.idxmin(), (skirtumas.min()).round(2), '%')
```
