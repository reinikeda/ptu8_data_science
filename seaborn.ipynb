```python
import pandas as pd
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
tips.shape
```

```python
sns.set_style('whitegrid')
```

```python
sns.displot(tips['total_bill'])
```

```python
sns.displot(tips['total_bill'], bins=10, kde=True)
```

```python
sns.scatterplot(tips, x='total_bill', y='tip', size='size')
```

```python
sns.color_palette('rocket')
```

```python
sns.jointplot(tips, x='total_bill', y='tip', kind='hex', cmap='rocket_r', color=sns.color_palette('rocket')[1])
```

```python
sns.jointplot(tips, x='total_bill', y='tip', hue='size', palette='coolwarm')
```

```python
sns.jointplot(tips, x='total_bill', y='tip', kind='reg', palette='coolwarm')
```

```python
sns.jointplot(tips, y='total_bill', x='tip', kind='kde', fill=True)
```

```python
sns.scatterplot(tips, x='total_bill', y='tip', size='size', hue='sex')
```

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips, hue='sex')
```

```python
sns.barplot(tips, x='sex', y='total_bill')
```

```python
sns.barplot(tips, x='sex', y='total_bill', estimator='sum')
```

```python
sns.barplot(tips, x='sex', y='total_bill', hue='day', estimator='sum')
```

```python
sns.countplot(tips, x='smoker')
```

```python
sns.countplot(tips, x='size')
```

```python
tips.tail()
```

```python
sns.boxplot(tips, x='smoker', y='total_bill')
```

```python
koreliacija = tips.corr(numeric_only=True)
koreliacija
```

```python
sns.heatmap(koreliacija, annot=True)
```

```python
fg = sns.FacetGrid(data=tips, col='time', row='smoker')
fg.map(sns.scatterplot, 'tip', 'total_bill')
```

```python
fg = sns.FacetGrid(data=tips, col='time', row='smoker')
fg.map(sns.countplot, 'size')
```
