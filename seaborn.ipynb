```python
import pandas as pd
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
tips.shape
```

## pasiskirstymo histogramos

```python
sns.set_style('whitegrid')
sns.displot(tips['total_bill'], bins=15, kde=True)
```

```python
sns.scatterplot(tips, x='total_bill', y='tip', size='size')
```

## .jointplot()

```python
sns.jointplot(tips, x='total_bill', y='tip', kind='hex', cmap='rocket_r', color=sns.color_palette('rocket')[2])
```

```python
sns.jointplot(tips, x='total_bill', y='tip', hue='sex')
```

```python
sns.jointplot(tips, x='total_bill', y='tip', hue='size', palette='coolwarm')
```

```python
sns.jointplot(tips, x='total_bill', y='tip', kind='reg')
```

```python
sns.jointplot(tips, x='total_bill', y='tip', kind='kde', fill=True)
```

### .scatterplot()

```python
sns.scatterplot(tips, x='total_bill', y='tip', size='size', hue='sex')
```

### .pairplot()

```python
sns.pairplot(tips, diag_kind='hist', hue='sex')
```

### kategorizavimo histogramos .barplot()

```python
sns.barplot(tips, x='sex', y='total_bill')
# vidurkis
```

```python
sns.barplot(tips, x='sex', y='total_bill', estimator=sum)
# suma
```

```python
sns.barplot(tips, x='sex', y='total_bill', hue='size', estimator=sum)
# suma pagal kompanijos dydi
```

### .countplot()

```python
sns.countplot(tips, x='size')
# susumuoja
```

## boxplot()

```python
sns.boxplot(tips, x='smoker', y='total_bill')
```

## .heatmap()

```python
koreliacija = tips.corr(numeric_only=True)
koreliacija
```

```python
sns.heatmap(koreliacija, annot=True)
```

### .FacetGrid()

```python
fg = sns.FacetGrid(data=tips, col='time', row='smoker')
```

```python
fg = sns.FacetGrid(data=tips, col='time', row='smoker')
fg.map(sns.scatterplot, 'tip', 'total_bill')
```

```python
fg = sns.FacetGrid(data=tips, col='time', row='smoker')
fg.map(sns.countplot, 'size')
```
