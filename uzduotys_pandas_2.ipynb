```python
import numpy as np
import pandas as pd
```

```python
top_left = pd.read_csv('text/top1-25-1.csv')
top_left
```

```python
top_right = pd.read_csv('text/top1-25-2.csv')
top_right
```

```python
bottom_left = pd.read_csv('text/top26-50-1.csv')
bottom_left
```

```python
bottom_right = pd.read_csv('text/top26-50-2.csv')
bottom_right
```

```python
bottom = pd.concat([bottom_left, bottom_right], axis=1)
bottom
```

```python
top = pd.merge(top_left, top_right, on=['Track.Name', 'Popularity'])
top
```

```python
full_data = pd.concat([top, bottom])
full_data.reset_index().drop(columns='index')
```

```python
# indeksas nuo 1
full_data.index = np.arange(1, len(full_data) +1)
full_data
```

```python
# Sukurkite grupavimo pagal žanrą objektą
genre_group = full_data.groupby('Genre')
genre_group
```

```python
# kokie žanrai lentelėje pasitaiko daugiau negu 3 kartus?
more_than_3 = genre_group.count()
more_than_3[more_than_3['Track.Name'] > 3][['Track.Name']]
```

```python
# Koks žanras pats populiariausias? Koks mažiausiai populiarus?
avg = genre_group['Popularity'].mean()
avg
```

```python
print(avg.idxmax(), avg.max())
print(avg.idxmin(), avg.min())
```

```python
# Sukurkite lentelę, kurioje matytųsi:
# koks žanras turi aukščiausią vidurkį kiekviename indikatoriuje
# bei pats vidurkis
```

```python
indicator = full_data.columns.tolist()[3:]
indicator
```

```python
genre = genre_group.mean().idxmax()
genre
```

```python
points = genre_group.mean().max()
points
```

```python
max_table = pd.DataFrame([indicator, genre, points], ['Indicator', 'Genre', 'Max points']).transpose()
max_table
```

```python
# Grįžkime prie sulipdytos lentelės. Ištraukite visas eilutes, kurios turi NaN reikšmių.
full_data.isnull().any()
```

```python
# Stulpelyje 'Genre' NaN reikšmes pakeiskite į 'pop'
full_data['Genre'].fillna('pop', inplace=True)
```

```python
# Stulpelyje 'Popularity' trūkstamas reikšmes pakeiskite į stulpelio vidurkį
full_data['Popularity'].fillna(value=full_data['Popularity'].mean(), inplace=True)
```
