### pixel manipulation

```python
doggy_pixels = doggy.getdata()
for pixel in doggy_pixels:
    print(pixel)
```

```python
changed_doggy_pixels = []
for pixel in doggy_pixels:
    if pixel[0] >= 180 and pixel[1] >= 180 and pixel[2] >=180:
        new_pixel = pixel[0] - 80, pixel[1] - 80, pixel[2] - 30
    else:
        new_pixel = pixel
    changed_doggy_pixels.append(new_pixel)
changed_doggy_pixels
```

```python
doggy.putdata(changed_doggy_pixels)
doggy
```

```python
from PIL import ImageEnhance
```

```python
doggy_contrast = ImageEnhance.Contrast(doggy)
doggy_contrast.enhance(1.5)
```

```python
doggy_contrast = ImageEnhance.Sharpness(doggy)
doggy_contrast.enhance(0.5)
```
