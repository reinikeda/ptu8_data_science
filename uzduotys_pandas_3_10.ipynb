```python
import numpy as np
import pandas as pd
```

```python
wiki = pd.read_html('https://lt.wikipedia.org/wiki/S%C4%85ra%C5%A1as:Lietuvos_miestai_pagal_gyventojus')
wiki
```

```python
cities = wiki[0]
cities.head()
```

```python
# Pakeiskite indeksą į stulpelį 'Miestas'
cities.set_index('Miestas').head()
```

```python
cities.drop(['Eilė', 'Tankumas (2019)'], axis=1, inplace=True)
```

```python
columns = cities.columns.tolist()
columns
```

```python
column_trim = []
for col in columns:
    trimmed = col.replace('\xa0m.', '')
    column_trim.append(trimmed)

cities.columns = column_trim
```

```python
cities.head()
```

```python
cities.fillna('0', inplace=True)
```

```python
bad_1897 = [not isinstance(x, int) for x in cities['1923']]
```

```python
cities['1923'].loc[bad_1897]
```

```python
cities.replace(['(??)', '(?)'], int(0), inplace=True)
```

```python
def fix_star(x):
    if type(x) == str and x[-1] == '*':
        return x[:-1]
    return x
```

```python
cities['1897'] = cities['1897'].apply(fix_star)
```

```python
cities[['1959', '1970', '1923', '1897']] = cities[['1959', '1970', '1923', '1897']].astype(np.int64)
```
